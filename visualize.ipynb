```python
import pandas as pd
import matplotlib.pyplot as plt

improved_models = ["Logistic Regression", "SVM", "Naive Bayes", "Neural Net",
                   "BERT Base", "BERT Large", "RoBERTa Base", "BART Base", "BART Large"]

accuracies = [0.9135, 0.9176, 0.9012, 0.9172, 0.9332, 0.9366, 0.9326, 0.9363, 0.9389]

# Creating a DataFrame for sorting
df = pd.DataFrame({'Model': improved_models, 'Accuracy': accuracies})

# Sorting by accuracy
df_sorted = df.sort_values(by='Accuracy')

# Creating the line plot with sorted values and larger font sizes
plt.figure(figsize=(14, 7))
plt.plot(df_sorted['Model'], df_sorted['Accuracy'], marker='o', color='purple', linestyle='-', linewidth=2, markersize=8)
plt.xlabel('Models', fontsize=14)
plt.ylabel('Accuracy', fontsize=14)
plt.title('Progression of Accuracy in Models for AG News Classification', fontsize=14)
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=10)
plt.ylim(0.88, 0.95)
plt.grid(True)

# Adding accuracy annotations with larger font
for i, txt in enumerate(df_sorted['Accuracy']):
    plt.annotate(f"{txt:.4f}", (df_sorted['Model'].iloc[i], df_sorted['Accuracy'].iloc[i]), 
                 textcoords="offset points", xytext=(0,10), ha='center', fontsize=10)

plt.show()
```

```python
# Separating data into basic methods and modern models
basic_models = improved_models[:4]
modern_models = improved_models[4:]

basic_accuracies = accuracies[:4]
modern_accuracies = accuracies[4:]

# Number of groups
n_groups = 4
```

```python
# Separating the models into basic and modern for coloring
basic_indices = df_sorted['Model'].isin(basic_models)
modern_indices = df_sorted['Model'].isin(modern_models)

# colors for the groups
color_basic = 'seagreen'
color_modern = 'darkorange'


# Creating the line plot with different colors for the model groups
plt.figure(figsize=(14, 7))

# Plotting basic methods in blue
plt.plot(df_sorted['Model'][basic_indices], df_sorted['Accuracy'][basic_indices], 
         marker='o', color=color_basic, linestyle='-', linewidth=2, markersize=8, label='Basic Methods')

# Plotting modern models in red
plt.plot(df_sorted['Model'][modern_indices], df_sorted['Accuracy'][modern_indices], 
         marker='o', color=color_modern, linestyle='-', linewidth=2, markersize=8, label='Transformer-based Models')

plt.xlabel('Models', fontsize=14)
plt.ylabel('Accuracy', fontsize=14)
plt.title('Progression of Accuracy in Models for AG News Classification', fontsize=14)
plt.xticks(rotation=45, fontsize=14)
plt.yticks(fontsize=12)
plt.grid(True)
plt.legend(fontsize=14)

# Adding accuracy annotations with larger font
for i, txt in enumerate(df_sorted['Accuracy']):
    plt.annotate(f"{txt:.4f}", (df_sorted['Model'].iloc[i], df_sorted['Accuracy'].iloc[i]), 
                 textcoords="offset points", xytext=(0,10), ha='center', fontsize=10)

plt.show()
```
